q1

```python
L = [11, 12, 13, 14]

# (i)
L.append(50)
L.append(60)
print("After adding 50 & 60:", L)

# (ii)
L.remove(11)
L.remove(13)
print("After removing 11 & 13:", L)

# (iii)
print("Ascending order:", sorted(L))

# (iv)
print("Descending order:", sorted(L, reverse=True))

# (v)
print("Is 13 in L?", 13 in L)

# (vi)
print("Count of elements:", len(L))

# (vii)
print("Sum of all elements:", sum(L))

# (viii)
odd_sum = sum([x for x in L if x % 2 != 0])
print("Sum of odd numbers:", odd_sum)

# (ix)
even_sum = sum([x for x in L if x % 2 == 0])
print("Sum of even numbers:", even_sum)

# (x) Sum of primes
def is_prime(n):
    if n < 2:
      return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

prime_sum = sum([x for x in L if is_prime(x)])
print("Sum of prime numbers:", prime_sum)

# (xi)
L.clear()
print("After clearing:", L)

# (xii)
del L
```

```python
D = {1: 5.6, 2: 7.8, 3: 6.6, 4: 8.7, 5: 7.7}

# (i)
D[8] = 8.8
print("After adding key=8:", D)

# (ii)
D.pop(2)
print("After removing key=2:", D)

# (iii)
print("key=6 present", 6 in D)

# (iv)
print("No. of elements:", len(D))

# (v)
print("Sum of values:", sum(D.values()))

# (vi)
D[3] = 7.1
print("Updated dict:", D)

# (vii)
D.clear()
print("Cleared dict:", D)
```

```python
S1 = {10, 20, 30, 40, 50, 60}
S2 = {40, 50, 60, 70, 80, 90}

# (i)
S1.update([55, 66])
print("After adding:", S1)

# (ii)
S1.discard(10)
S1.discard(30)
print("After removing:", S1)

# (iii)
print("40 in S1", 40 in S1)

# (iv)
print("Union:", S1 | S2)

# (v)
print("Intersection:", S1 & S2)

# (vi)
print("Difference S1-S2:", S1 - S2)
```
